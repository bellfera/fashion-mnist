```python
# Normalizar entre 0 y 1
train_images = train_images / 255.0
# train_images /= 255.0
test_images = test_images / 255.0
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(512, activation='relu', kernel_initializer='he_uniform'),
    tf.keras.layers.Dense(10, activation='softmax')

])
```

```python
model.summary()
```

```python
# Función loss
# Clasificación binaria: binary_crossentropy
# Clasificación multiclase: sparse_categorical_crossentropy
# Genérica (siempre funciona bastante): mse
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(train_images, train_labels, batch_size=64, epochs=20)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=0)
print('Test accuracy:', test_acc)
print('Test loss:', test_loss)
```

```python
# Nuevo modelo usando técnica dropout
model2 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(784, activation='relu', kernel_initializer='he_uniform'),
    tf.keras.layers.Dropout(0.25),
    tf.keras.layers.Dense(10, activation='softmax')

])
```

```python
model2.summary()
```

```python
# Entrenamiento con dropout y early stopping
from tensorflow.keras.callbacks import EarlyStopping
es = EarlyStopping(monitor='val_accuracy', mode='max', verbose=1, patience=5, restore_best_weights=True)

model2.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
from sklearn.model_selection import train_test_split

# Separa en entrenamiento y validación (por ejemplo 80% train, 20% val)
X_train, X_val, y_train, y_val = train_test_split(
    train_images, train_labels,
    test_size=0.2,       # 20% de los datos para validación
    random_state=42,     # semilla para reproducibilidad
    shuffle=True         # mezclar antes de dividir
)

print("X_train:", X_train.shape)
print("X_val:", X_val.shape)
print("y_train:", y_train.shape)
print("y_val:", y_val.shape)
```

```python
model2.fit(X_train, y_train, batch_size=64, epochs=100, callbacks=[es], validation_data=(X_val, y_val))
```

```python
test_loss, test_acc = model2.evaluate(test_images, test_labels, verbose=0)
print('Test accuracy:', test_acc)
print('Test loss:', test_loss)
```
